**Ejemplo de Crawling **

```
import urllib
from bs4 import BeautifulSoup

url = 'https://www.jumbo.com.ar/'
html = urllib.request.urlopen(url)
soup = BeautifulSoup(html)

tags = soup('a')

no_es_enlace=[]
```

```
for tag in tags:
  print(tag.get('href'))

print('\r\n Enlaces de las páginas secundarias: \r\n')
```

```
for tag in tags:
	newurl = tag.get('href')
	print('* Accediendo a los enlaces dentro de la página',newurl)
	try:
		if newurl[0:4]=='http' :html2 = urllib.request.urlopen(str(newurl))
		else: html2 = urllib.request.urlopen(url + newurl)
	except: 
		print("¡Me quiero volver chango!")
		no_es_enlace.append(newurl)
```
